```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import os
from sklearn.base import BaseEstimator, TransformerMixin
from classes_for_pipeline_py import FeaturesAdder, DataDropper, CappedTargetDropper, small_Pipeline

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import PowerTransformer, StandardScaler

from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import mean_absolute_error as MAE
from sklearn.metrics import get_scorer

from sklearn.linear_model import LinearRegression, HuberRegressor, Lasso, Ridge
from sklearn.model_selection import cross_val_score, train_test_split, learning_curve, ShuffleSplit, KFold

# to supress "UserWarning: X does not have valid feature names, but PowerTransformer was fitted with feature names"
import warnings
warnings.filterwarnings("ignore")

np.set_printoptions(suppress=True)

RMSE = lambda y_true, y_test: np.sqrt(MSE(y_true, y_test))
```

```python
HOUSING_DATASET_CSV = os.path.join(os.getcwd(), 'housing_dataset\housing_data.csv')
```

```python
housing_dataset = pd.read_csv(HOUSING_DATASET_CSV)
print(housing_dataset.shape)
housing_dataset.head()
```

```python
# split the dataset
train, test = train_test_split(housing_dataset, test_size=0.2, random_state=42)

# split the training
X_train_pp, X_test_pp = train_test_split(train, test_size=0.2, random_state=42)
```

```python
class PreprocessPipeline(BaseEstimator, TransformerMixin):
    
    '''any transformer can be access via 
    PreprocessPipeline.name_of_transformer.list_of_transformers[0][0].statistic_in_question'''
    
    def __init__(self):
        self.median_imputer = small_Pipeline([
            (SimpleImputer(strategy='median'), 
             ['longitude', 'latitude', 'housing_median_age',
              'total_rooms', 'total_bedrooms', 'population', 
              'households', 'median_income', 'median_house_value'])
        ])
        self.feature_adder = FeaturesAdder()
        self.outliers_dropper = small_Pipeline([
            (CappedTargetDropper(capped_val=500000.0), 'median_house_value'),
            (DataDropper(method='optimized', penalty=0.25), 'income_per_household'),
            (DataDropper(method='optimized', penalty=0.25), 'population_per_household'),
            (DataDropper(method='optimized', penalty=0.25), 'rooms_per_household'),
            (DataDropper(method='optimized', penalty=0.25), 'rooms_per_age')
        ])
        self.standardizer = small_Pipeline([
            (StandardScaler(), 
             ['longitude', 'latitude'])
        ])
        self.features_power_transformer = small_Pipeline([
            (PowerTransformer(),
             ['housing_median_age', 'total_rooms',
              'total_bedrooms', 'population', 'households', 'median_income',
              'rooms_per_household', 'income_per_household',
              'income_per_population', 'bedrooms_per_rooms',
              'population_per_household', 'rooms_per_age'])
        ])
        self.target_power_transformer = small_Pipeline([
            (PowerTransformer(), ['median_house_value'])
        ])
        
    def fit(self, X):
        
        self.median_imputer.fit(X)
        
        # since all the later methods are obtained with nans imputed and added
        # new features we have to transform X and add new features,
        # then fit the other classes
        X = self.median_imputer.transform(X)
        X = self.feature_adder.fit_transform(X)
        
        self.outliers_dropper.fit(X)
        self.standardizer.fit(X)
        self.features_power_transformer.fit(X)
        self.target_power_transformer.fit(X)
        
        return self
    
    def transform(self, X):
        
        X = self.median_imputer.transform(X)
        X = self.feature_adder.fit_transform(X)
        X = self.outliers_dropper.fit_transform(X)
        X = X.dropna()
        X = self.standardizer.transform(X)
        X = self.features_power_transformer.transform(X)
        X = self.target_power_transformer.transform(X)
        
        X_cat = X['ocean_proximity'].copy()
        X_num = X.drop(columns=['ocean_proximity'])
        
        X_cat.loc[X_cat == 'ISLAND'] = np.nan
        X_merged = pd.merge(X_num, 
                            pd.get_dummies(X_cat, drop_first=True), 
                            left_index=True, right_index=True)
        
        return X_merged

PP = PreprocessPipeline()
```

```python
# fit and transform the data
PP.fit(X_train_pp)
X_train, X_test = PP.transform(X_train_pp), PP.transform(X_test_pp)

# split to features and target
X_train, X_test, y_train, y_test = X_train.drop(columns=['median_house_value']).values,\
                                   X_test.drop(columns=['median_house_value']).values,\
                                   X_train.loc[:, 'median_house_value'].values,\
                                   X_test.loc[:, 'median_house_value'].values
```

```python
def metrics_summary(model, X, y):
    list_of_metrics = ['neg_root_mean_squared_error', 'neg_mean_absolute_error']
    for metric, name in zip(list_of_metrics, ['RMSE', 'MAE']):
        cv_score = -cross_val_score(model, X, y, scoring=metric, cv=5)
        print(f'{name}: {cv_score.mean().round(2)} +- {(2 * cv_score.std() / np.sqrt(5)).round(2)}')

def target_power_inverse(pipeline, target):
    '''
    The function inverses the predicted target, if the target was choosen to be
    power transformed.
    '''
    inv_target = pipeline.target_power_transformer\
                         .list_of_transformers[0][0]\
                         .inverse_transform(
                             target.reshape(-1, 1)
                         )
    return inv_target

def plot_learning_curves(model, X, y, metric='neg_root_mean_squared_error', train_size=np.linspace(0.05, 1, 5), cv=5):
    S, train_scores, test_scores = learning_curve(model, X, y, scoring=metric, train_sizes=train_size, cv=cv, shuffle=True)
    train_scores, test_scores = -train_scores, -test_scores
    
    plt.plot(S, test_scores.mean(axis=1), color='black', lw=0.5)
    plt.fill_between(S, 
                     test_scores.mean(axis=1) + 2 * test_scores.std(axis=1) / np.sqrt(cv), 
                     test_scores.mean(axis=1) - 2 * test_scores.std(axis=1) / np.sqrt(cv), 
                     color='black', alpha=0.1)
    
    plt.plot(S, train_scores.mean(axis=1), color='red', lw=0.5)
    plt.fill_between(S, 
                     train_scores.mean(axis=1) + 2 * train_scores.std(axis=1) / np.sqrt(cv), 
                     train_scores.mean(axis=1) - 2 * train_scores.std(axis=1) / np.sqrt(cv), 
                     color='red', alpha=0.1)
    
    plt.axhline(
        np.mean(
            [test_scores.mean(axis=1)[-1], 
             train_scores.mean(axis=1)[-1]]
        ), color='grey', ls='--')
    plt.show()
    
def cv_with_inverse(model, X, y, metric, n_splits=10, test_size=0.2, random_state=None):
    
    cv = ShuffleSplit(n_splits=n_splits, test_size=test_size, random_state=random_state)
    
    tmp_scores = np.empty((0))
    for train_idx, test_idx in cv.split(X):
        
        X_train_tmp, X_test_tmp, y_train_tmp, y_test_tmp = X[train_idx, :], \
                                                           X[test_idx, :], \
                                                           y[train_idx], \
                                                           y[test_idx]
        
        model.fit(X_train_tmp, y_train_tmp)
        
        y_pred = model.predict(X_test_tmp)
        
        # inverse
        y_test_inv = target_power_inverse(PP, y_test_tmp)
        y_pred_inv = target_power_inverse(PP, y_pred)
        
        # store the values
        score = metric(y_test_inv, y_pred_inv)
        tmp_scores = np.r_[tmp_scores, score]
        
    return tmp_scores

def learning_curves_with_inverse(model, X, y, 
                                 metric=RMSE, train_sizes=np.linspace(0.05, 1, 5), 
                                 cv=5, shuffle=False, random_state=None):
    
    # create a list of tuples containing the train and test indices
    cv_list = list(KFold(n_splits=cv, shuffle=shuffle, random_state=random_state).split(X))
    
    # we are interested in the size of any train
    max_samples = cv_list[0][0].size
    
    # if train_sizes are in % change them into number of samples
    if train_sizes.dtype == 'float':
        train_sizes = (train_sizes * max_samples).astype(int)
    
    if np.max(train_sizes) > max_samples:
        print('The maximum value of samples if greater than the size of \
              cross-validation training set. The maximum is set to {max_samples}')
    
    train_sizes = np.clip(train_sizes, 1, max_samples)
    
    train_scores = np.zeros((train_sizes.size, cv))
    test_scores = np.zeros_like(train_scores)
    
    for i, n_sample in enumerate(train_sizes):
        
        for j, (train, test) in enumerate(cv_list):
            
            train = train[:n_sample]
            
            _X_train, _X_test, _y_train, _y_test = X[train], X[test],\
                                                   y[train], y[test]
            
            model.fit(_X_train, _y_train)
            
            y_pred_train = model.predict(_X_train)
            y_pred_test = model.predict(_X_test)
            
            # inverse
            y_train_inv = target_power_inverse(PP, _y_train)
            y_test_inv = target_power_inverse(PP, _y_test)
            y_pred_train_inv = target_power_inverse(PP, y_pred_train)
            y_pred_test_inv = target_power_inverse(PP, y_pred_test)
            
            # store the values           
            train_scores[i, j] = metric(y_train_inv, y_pred_train_inv)
            test_scores[i, j] = metric(y_test_inv, y_pred_test_inv)
            
    return train_sizes, train_scores, test_scores

def plot_learning_curves_inverse(model, X, y, 
                                 metric=RMSE, train_sizes=np.linspace(0.05, 1, 5), 
                                 cv=5, shuffle=False, random_state=None):
    
    S, train_scores, test_scores = learning_curves_with_inverse(model, X, y, 
                                                                metric=RMSE, train_sizes=train_sizes, 
                                                                cv=cv, shuffle=shuffle, random_state=random_state
                                                               )
    
    plt.plot(S, test_scores.mean(axis=1), color='black', lw=0.5)
    plt.fill_between(S, 
                     test_scores.mean(axis=1) + 2 * test_scores.std(axis=1) / np.sqrt(cv), 
                     test_scores.mean(axis=1) - 2 * test_scores.std(axis=1) / np.sqrt(cv), 
                     color='black', alpha=0.1)
    
    plt.plot(S, train_scores.mean(axis=1), color='red', lw=0.5)
    plt.fill_between(S, 
                     train_scores.mean(axis=1) + 2 * train_scores.std(axis=1) / np.sqrt(cv), 
                     train_scores.mean(axis=1) - 2 * train_scores.std(axis=1) / np.sqrt(cv), 
                     color='red', alpha=0.1)
    
    plt.axhline(
        np.mean(
            [test_scores.mean(axis=1)[-1], 
             train_scores.mean(axis=1)[-1]]
        ), color='grey', ls='--')
    plt.show()
```

```python
# Classical Linear Regression
LR = LinearRegression()
LR.fit(X_train, y_train)
y_pred_LR = LR.predict(X_test)
```

```python
plot_learning_curves_inverse(LR, X_train, y_train, train_sizes=np.linspace(0.05, 1, 10), cv=5)
```
